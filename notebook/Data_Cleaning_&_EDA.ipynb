```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.xls', header=1)
```

```python
df.head()
```

```python
df.info()
```

### Data Cleaning

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122, 'Region'] = 0
df.loc[122:, 'Region'] = 1
```

```python
df['Region'] = df['Region'].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.loc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.loc[[122]]
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns
```

```python
int_cols = ['day', 'month', 'year', 'Temperature', 'RH', 'Ws']
df[int_cols] = df[int_cols].astype(int)
```

```python
float_cols = ['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']
df[float_cols] = df[float_cols].astype(float)
```

```python
df.info()
```

```python
df['Classes'].value_counts()
```

```python
def classes_clear(txt):
    if 'not' in txt:
        return 'not fire'
    else:
        return 'fire'
df['Classes'] = df['Classes'].apply(classes_clear)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```
